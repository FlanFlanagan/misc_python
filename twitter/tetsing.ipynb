```python
import numpy as np
import scipy as sp
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
x = pd.read_csv('original.csv')
t1 = pd.to_datetime(x['created_at'])
sub = (t1.dt.day*60*24+t1.dt.hour*60+t1.dt.minute)[0]

x = pd.read_csv('level1.csv')
t1 = pd.to_datetime(x['timestamp'])
x['minute'] = ((t1.dt.day*60*24+t1.dt.hour*60+t1.dt.minute)-sub)
x['hour'] = ((t1.dt.day*60*24+t1.dt.hour*60+t1.dt.minute)-sub)/60
x['second'] = ((t1.dt.day*60*24+t1.dt.hour*60+t1.dt.minute)*60+t1.dt.second)-sub*60

y = np.array(x['minute'])
plt.hist(y, bins='fd')
```

```python
h = np.arange(0,5000, 30)
y = np.array(x['minute'])
plt.hist(y, bins=h)
```
