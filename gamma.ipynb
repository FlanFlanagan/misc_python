```python
import numpy as np
import matplotlib.pyplot as plt
import copy
```

```python
def gamma(rate, events, rrange):
    y = copy.deepcopy(rrange)
    p = []
    for t in y:
        g = (rate*np.exp(-rate*t))
        g *= (rate*t)**(events-1)
        g /= np.math.factorial(events-1)
        p.append(g)
    p = np.array(p)
    return p

def gamma_t(rate, events, t):
    g = (rate*np.exp(-rate*t))
    g *= (rate*t)**(events-1)
    g /= np.math.factorial(events-1)
    return g
```

```python
x = np.linspace(0, 10, 1000)
for i in range(1, 7, 2):
    y = gamma(1, i, x)
    lstr = str(i) + ' events'
    plt.plot(x, y, label=lstr)
plt.xlabel('Time(unit of time associated with rate)')
plt.ylabel('f(t)')
plt.legend()
```

```python
x = np.linspace(0, 4, 1000)
for i in range(2, 12, 2):
    y = gamma(2, i, x)
    lstr = str(i) + ' events'
    plt.plot(x, y, label=lstr)
plt.xlabel('Time(unit of time associated with rate)')
plt.ylabel('f(t)')
plt.legend()
```

```python
x = np.linspace(0, 4, 1000)
y = gamma(2, 4, x)
lstr = '2 fish/hour'
plt.plot(x, y, label=lstr)
y = gamma(0.5, 4, x)
lstr = '0.5fish/hour'
plt.plot(x, y, label=lstr)

plt.xlabel('Time(h)')
plt.ylabel('f(t)')
plt.legend()
```
