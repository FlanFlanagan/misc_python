```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import random
import pandas as pd
import math
```

```python
draws = 100
for j in np.arange(0.1, 0.6, 0.1):
    m = []
    for i in range(draws):
        m.append(stats.binom.pmf(i, draws,  j))
    plt.plot(m, label='P_success =' +str(j))
    plt.xlabel('Number of Successes')
    plt.ylabel('Probability')
    plt.legend()
```

```python
mu = 0
for i in range(1, 4, 1):
    variance = i
    sigma = math.sqrt(variance)
    x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
    plt.plot(x, stats.norm.pdf(x, mu, sigma), label='sigma = '+str(i))
plt.xlabel('Value')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

```python
mu = 0
for i in range(1, 4, 1):
    variance = i
    sigma = math.sqrt(variance)
    x = np.linspace(-5, 5, 100)
    plt.plot(x, stats.norm.cdf(x, mu, sigma), label='sigma = '+str(i))
plt.xlabel('Value')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

```python
def poisson(lam, k):
    x = lam**k * (np.exp(-lam))/math.factorial(k)
    return x

lam = 5
vals = []
x = range(0,10,1)
print(x)
for i in x:
    vals.append(poisson(lam, i))

plt.plot(x, vals)
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Probability of <x> events in 1 year lambda = '+str(lam))
plt.show()
```

```python
def gamma_dist(x, k, theta):
    c = 1/(math.factorial(k-1) * theta**k)
    x_fact = x**(k-1)
    exp = np.exp(-1*x/theta)
    return c * x_fact * exp

x = np.arange(0, 1, 0.00001)
y = gamma_dist(x, 2, 1)
plt.plot(x, y, label='Test')
y.sum()*0.00001
```

```python
def power_dist(x, c, k):
    v = c * x**k
    return v

x = np.arange(0.01, 1, 0.01)
y = power_dist(x, 1, -1.01)

plt.plot(x, y)
plt.title('Power Distribution; c=1, k=-2')
plt.ylabel('Frequency')
```
