```python
draws = 100
for j in np.arange(0.1, 0.6, 0.1):
    m = []
    for i in range(draws):
        m.append(stats.binom.pmf(i, draws,  j))
    plt.plot(m, label='P_success =' +str(j))
    plt.xlabel('Number of Successes')
    plt.ylabel('Probability')
    plt.legend()
```

```python
mu = 0
for i in range(1, 4, 1):
    variance = i
    sigma = math.sqrt(variance)
    x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
    plt.plot(x, stats.norm.pdf(x, mu, sigma), label='sigma = '+str(i))
plt.xlabel('Value')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

```python
def poisson(lam, k):
    x = lam**k * (np.exp(-lam))/math.factorial(k)
    return x

lam = 5
vals = []
x = range(0,10,1)
print(x)
for i in x:
    vals.append(poisson(lam, i))

plt.plot(x, vals)
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Probability of <x> events in 1 year lambda = '+str(lam))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def exponential_dist(x, lamb):
    return lamb*np.exp(-1*lamb*x)

geys = [19, 20, 21, 21, 4, 34, 41, 20, 9, 11, 5, 21, 7, 40, 32]
geys = np.array(geys)/60
lamb = geys.mean()

x = np.arange(0, 5, 0.1)
y = []
for p in x:
    y.append(exponential_dist(p, lamb))
plt.plot(x, y, label='New Data')

y2 = []
for p in x:
    y2.append(exponential_dist(p, (17.5/60)))
plt.plot(x, y2, label='Old Data')
```

```python
def gamma(x, k, theta):
    return 1/(math.factorial(k-1) * (theta**k)) * x**(k-1) * np.exp(-1* (x/theta))

geys = [19, 20, 21, 21, 4, 34, 41, 20, 9, 11, 5, 21, 7, 40, 32]
geys = np.array(geys)
lamb = geys.mean()

x = np.arange(0, 200, 0.1)
y = []
for p in x:
    y.append(gamma(p, 2, lamb))
plt.plot(x, y, label='New Data')
```
