```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import random
import pandas as pd
import math
```

```python
stats.binom.stats(8, 0.4)
```

```python
draws = 100
for j in np.arange(0.1, 0.6, 0.1):
    m = []
    for i in range(draws):
        m.append(stats.binom.pmf(i, draws,  j))
    plt.plot(m, label='P_success =' +str(j))
    plt.xlabel('Number of Successes')
    plt.ylabel('Probability')
    plt.legend()
```

```python
mu = 0
for i in range(1, 4, 1):
    variance = i
    sigma = math.sqrt(variance)
    x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
    plt.plot(x, stats.norm.pdf(x, mu, sigma), label='sigma = '+str(i))
plt.xlabel('Value')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

```python
def poisson(lam, k):
    x = lam**k * (np.exp(-lam))/math.factorial(k)
    return x

lam = 5
vals = []
x = range(0,10,1)
print(x)
for i in x:
    vals.append(poisson(lam, i))

plt.plot(x, vals)
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Probability of <x> events in 1 year lambda = '+str(lam))
plt.show()
```
